Target-Rate per Category plots.

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
cat_cols += [col for col in df.columns if 'cat' in col or 'bin' in col ]

cat_cols = list(dict.fromkeys(cat_cols))  #REMOVE DUPLICATES.
for col in cat_cols:
    target_rate = df.groupby(col)['target'].mean().sort_values()

    plt.figure(figsize=(8,4))
    sns.pointplot(x=target_rate.index, y=target_rate.values)
    plt.title(f"Target Rate by {col}")
    plt.xlabel(col)
    plt.ylabel(f"Mean {'target'}")
    plt.xticks(rotation=360)
    plt.show()

```

```python
def target_rate_summary(df,target_col='target'):
    cat_cols = df.select_dtypes(include=['object']).columns.tolist()
    cat_cols += [col for col in df.columns if 'cat' in col or 'bin' in col]
    cat_cols = list(dict.fromkeys(cat_cols))  # remove duplicates

    summary_rows = []

    for col in cat_cols:
        rates = df.groupby(col)[target_col].mean()
        counts = df.groupby(col)[target_col].count()

        best_cat = rates.idxmin()
        worst_cat = rates.idxmax()

        summary_rows.append({
            "Feature": col,
            "Best_Category": best_cat,
            "Best_Target_Rate": rates[best_cat],
            "Best_Count": counts[best_cat],

            "Worst_Category": worst_cat,
            "Worst_Target_Rate": rates[worst_cat],
            "Worst_Count": counts[worst_cat]
        })

    summary_df = pd.DataFrame(summary_rows)
    return summary_df


summary = target_rate_summary(df, target_col='target')
summary
```

**SPLITTING DATA**

```python
#splitting the DATA into training data and testing data
from sklearn.model_selection import train_test_split
labels = df.columns[2:]
df_train = df.dropna()

x = df_train[labels]
y = df_train['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**SCALING DATA**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# APPLY RANDOM UNDERSAMPLING,CLUSTER CENTROID

**Random Undersampling**

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()
x_rus,y_rus = rus.fit_resample(x_train_scaled,y_train)
print("before random sampling",y_train.value_counts())
print("after random sampling",pd.Series(y_rus).value_counts())
```

AFTER SCALING THE DATA


Though the data is properly balanced into 1:1 ratio,

i.e.

0 -- 17309

1 -- 17309

458k samples from class 0, may loose a massive amount of information like, important patterns in the majority class,



**Cluster Centroids**

```python
from imblearn.under_sampling import ClusterCentroids
cc = ClusterCentroids(sampling_strategy={0:200},random_state=42)
X_cc, y_cc = cc.fit_resample(x_train_scaled, y_train)

print("\nBefore ClusterCentroids:", y_train.value_counts())
print("After ClusterCentroids:\n", pd.Series(y_cc).value_counts())
```

**ML model fitting**

ON DATA AFTER RANDOM UNDER SAMPLING

```python
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

models = {
    'Logistic Regression': LogisticRegression(class_weight='balanced'),
    'Decision Tree': DecisionTreeClassifier()
}

print("\n===== UNDERSAMPLING RESULTS ====")
for name, model in models.items():
    model.fit(x_rus, y_rus)
    y_pred = model.predict(x_test_scaled)
    y_prob = model.predict_proba(x_test_scaled)[:, 1]

    print("\n Model:", name)
    print("F1 Score:", f1_score(y_test, y_pred))
    print("AUC Score:", roc_auc_score(y_test, y_prob))
```

ON DATA AFTER APPLYING CLUSTER CENTROIDS

# RANDOM OVERSAMPLING AND SMOTE

Random OverSampling

```python
#performing Random Over-Sampling to balance the minority class

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
x_ros,y_ros = ros.fit_resample(x_train_scaled,y_train)
print("before",y.value_counts())
print("after",pd.Series(y_ros).value_counts())
```

**SMOTE**(Synthetic Minority Oversampling TEchnique)



```python
from imblearn.over_sampling import SMOTE
import pandas as pd

sm = SMOTE(random_state=42)
x_sm, y_sm = sm.fit_resample(x_train_scaled, y_train)

print("\nBefore SMOTE:", y.value_counts())
print("After SMOTE:\n", pd.Series(y_sm).value_counts())
```

**ML model fitting**

ON DATA AFTER APPLYING RANDOM OVERSAMPLING

```python
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

models = {
    'Logistic Regression': LogisticRegression(class_weight='balanced'),
    'Decision Tree': DecisionTreeClassifier()
}

print("\n===== OVERSAMPLING RESULTS ====")
for name, model in models.items():
    model.fit(x_ros,y_ros)
    y_pred = model.predict(x_test_scaled)
    y_prob = model.predict_proba(x_test_scaled)[:, 1]

    print("\n Model:", name)
    print("F1 Score:", f1_score(y_test, y_pred))
    print("AUC Score:", roc_auc_score(y_test, y_prob))
```
